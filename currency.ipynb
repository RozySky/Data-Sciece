```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
src = requests.get("https://www.exchangerates.org.uk/USD-PKR-exchange-rate-history.html").content
soup = BeautifulSoup(src, 'lxml')
list_rates = soup.find_all('td')
rates = []
for items in list_rates:
    if '=' in items.text:
        rates.append(float(items.text.split('1 USD = ')[-1].replace('PKR', '').strip()))
PKR = rates[0:30]
PKR.reverse()
```

```python
src2 = requests.get("https://www.exchangerates.org.uk/USD-INR-exchange-rate-history.html").content
soup = BeautifulSoup(src2, 'lxml')
list_rates = soup.find_all('td')
rates = []
for items in list_rates:
    if '1 USD =' in items.text:
        rates.append(float(items.text.split('1 USD = ')[-1].replace('INR', '').strip()))
INR = rates[0:30]
INR.reverse()
```

```python
days = [i for i in range(1,31)]
```

```python
plt.plot(days,INR, 'r')
plt.xlabel("Days in Month")
plt.ylabel("Dollar Rate in INR")
```

```python
plt.plot(days,PKR,)
plt.xlabel("Days in Month")
plt.ylabel("Dollar Rate in PKR")
```

```python
names = ['India INR', 'Pakistan PKR']
average = [np.mean(INR), np.mean(PKR)]
plt.bar(names,average)
plt.suptitle("Increase In rate of Dollar In INR & PKR.")
```
